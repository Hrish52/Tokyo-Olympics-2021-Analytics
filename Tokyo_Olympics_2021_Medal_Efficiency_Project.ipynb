## Tokyo Olympics 2021 - Medal Efficiency Project

### Description


This project delves into the true performance of nations at the Tokyo Olympics 2021 by evaluating their medal efficiency. Beyond raw medal counts, we analyze how effectively each country converted its resources into Olympic success.

By comparing medal hauls against factors like the number of participating athletes, population, and GDP per capita, this project highlights countries that truly "punch above their weight" on the global sporting stage. Key metrics explored include:

*   **Medals per Athlete**: To understand the productivity of each athlete in a country's delegation.
*   **Medals per Capita**: To gauge a nation's sporting prowess relative to its population size.
*   **Medals per $B GDP**: To assess efficiency in medal acquisition concerning economic power.

This analysis provides a nuanced perspective on Olympic success, revealing the nations with exceptional athletic programs and resource utilization.

##### Cross-Table Country Validation

Identify discrepancies in country names across the Athletes, Teams, and Medals datasets and identify countries with zero medals.

We will perform a set-based comparison between the 'NOC' columns in Athletes and Teams and the 'Team/NOC' column in Medals to identify zero-medal nations and potential naming discrepancies.

```python
# 1. Extract unique country names and convert to sets
athletes_countries = set(dfs['Athletes']['NOC'].unique())
teams_countries = set(dfs['Teams']['NOC'].unique())
medals_countries = set(dfs['Medals']['Team/NOC'].unique())

# 2. Find countries present in Athletes or Teams but NOT in Medals (Zero-medal nations)
participating_countries = athletes_countries | teams_countries
zero_medal_countries = participating_countries - medals_countries

# 3. Identify naming mismatches (In Medals but not in Athletes)
medals_not_in_athletes = medals_countries - athletes_countries

# 4. Print results
print(f"Total unique countries in Athletes: {len(athletes_countries)}")
print(f"Total unique countries in Teams: {len(teams_countries)}")
print(f"Total unique countries in Medals: {len(medals_countries)}")
```

```python
print(f"\n--- Zero-Medal Countries ({len(zero_medal_countries)}) ---")
print(sorted(list(zero_medal_countries)))

print(f"\n--- Naming Discrepancies (In Medals but not Athletes) ({len(medals_not_in_athletes)}) ---")
if not medals_not_in_athletes:
    print("None found. Country names in Medals match those in Athletes.")
else:
    print(sorted(list(medals_not_in_athletes)))
```

Reconcile athlete counts from the Athletes table with EntriesGender and Teams tables, investigate Coaches duplicates, and identify small delegations.


We will calculate the athlete counts per NOC, verify the total entries from EntriesGender against the Athletes DataFrame, check for duplicates in Coaches, and identify small delegations as requested in the instructions.

```python
# 1. Calculate the total number of athletes per NOC
athletes_per_noc = dfs['Athletes']['NOC'].value_counts().reset_index()
athletes_per_noc.columns = ['NOC', 'Athlete_Count']

# 2. Reconcile athlete counts with EntriesGender
total_athletes_df = len(dfs['Athletes'])
total_entries_sum = dfs['EntriesGender']['Total'].sum()
discrepancy = total_athletes_df - total_entries_sum

# 3. Check for duplicates in Coaches
coaches_duplicates = dfs['Coaches'].duplicated().sum()

# 4. Identify small delegations (1-3 athletes)
small_delegations = athletes_per_noc[athletes_per_noc['Athlete_Count'] <= 3]
```

#### STEP 1: Core Efficiency Metrics

These are the main metrics for your dashboard.For each one, think about what story it tells.

- Medals per Athlete
    → How efficiently did a country convert athletes into medals?
    
    TODO: final_df['Medals_Per_Athlete'] = ...
    Hint: Total medals / Athlete_Count


- Medals per Million Population
    → Small countries with medals are punching above their weight
    
    TODO: final_df['Medals_Per_Million_Pop'] = ...
    Hint: (Total medals / Population) * 1_000_000


- Medals per Billion USD GDP
    → Are richer countries just buying medals with investment?
    
    TODO: final_df['Medals_Per_Billion_GDP'] = ...
    Hint: (Total medals / GDP) * 1_000_000_000


- Gold Medal Percentage

    → What proportion of a country's medals are gold?

    → A country with 5 golds out of 5 medals is more dominant than one with 5 golds out of 50

    TODO: final_df['Gold_Percentage'] = ...
    Hint: (Gold / Total) * 100
    Watch out: avoid division by zero if Total is 0

```python
# Load the final dataframe from dfs
final_df = dfs['Final_Analysis_Data']

# 1. Medals per Athlete
final_df['Medals_Per_Athlete'] = final_df['Total'] / final_df['Athlete_Count']

# 2. Medals per Million Population
# Population is in absolute numbers, so divide by 1,000,000 for 'per million'
final_df['Medals_Per_Million_Pop'] = (final_df['Total'] / final_df['Population']) * 1_000_000

# 3. Medals per Billion USD GDP
# GDP is typically in USD, so divide by 1,000,000,000 for 'per billion'
final_df['Medals_Per_Billion_GDP'] = (final_df['Total'] / final_df['GDP']) * 1_000_000_000

# 4. Gold Medal Percentage
# Handle division by zero for countries with 0 total medals (though we filtered for > 0 medals, it's good practice)
final_df['Gold_Percentage'] = (final_df['Gold'] / final_df['Total']) * 100
final_df['Gold_Percentage'] = final_df['Gold_Percentage'].fillna(0) # If Total is 0, percentage is 0

# Update the stored dataframe
dfs['Final_Analysis_Data'] = final_df
```

#### STEP 2: Weighted Medal Score

Not all medals are equal — a gold is worth more than a bronze. Create a weighted score to reflect this.

Common weighting: Gold = 3, Silver = 2, Bronze = 1 (Some analysts use 5-3-1 or 4-2-1 — pick one and justify it)

TODO: final_df['Weighted_Medal_Score'] = ...
Hint: (Gold * 3) + (Silver * 2) + (Bronze * 1)

Then create weighted efficiency metrics:

TODO: final_df['Weighted_Score_Per_Athlete'] = ...

TODO: final_df['Weighted_Score_Per_Million_Pop'] = ...

```python
final_df = dfs['Final_Analysis_Data']

# 1. Weighted Medal Score (Gold=3, Silver=2, Bronze=1)
final_df['weighted_medal_score'] = (final_df['gold'] * 3) + (final_df['silver'] * 2) + (final_df['bronze'] * 1)

# 2. Weighted Score Per Athlete
final_df['weighted_score_per_athlete'] = final_df['weighted_medal_score'] / final_df['athlete_count']

# 3. Weighted Score Per Million Population
final_df['weighted_score_per_million_pop'] = (final_df['weighted_medal_score'] / final_df['population']) * 1_000_000

# Update the stored dataframe
dfs['Final_Analysis_Data'] = final_df

print("Weighted medal scores and weighted efficiency metrics calculated.")
print("Preview of final_df with new weighted metrics:")
display(final_df.head())
```
